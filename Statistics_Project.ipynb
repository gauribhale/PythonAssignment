```python
import pandas as pd
import numpy as np
import seaborn as sns
```

1. Reading thhe dataset

```python
data=pd.read_csv('E:\DataScience\DataSets\Laliga.csv')
data.head()
```

2. Renaming column names by first row of the data set

```python
data.rename(columns=data.iloc[0],inplace=True)
```

```python
data.head()
```

3. Dropping the first row

```python
data.drop(data.index[0],inplace=True)
```

```python
data.head()
```

4. Identifying the data types

```python
data.info()
```

5. Replacing dash with 0

```python
data.replace('-',0,inplace=True)
```

```python
data.head()
```

6. Finding teams whose debut is between year 1930 and 1980

```python
data['Debut']=data['Debut'].astype('str').str[:4].astype(int)
```

```python
data_new=data[data['Debut'].between(1930,1980)]
data_new.head()
```

```python
data['Points']=data['Points'].apply(pd.to_numeric)
data['GoalsFor']=data['GoalsFor'].apply(pd.to_numeric)
data['GoalsAgainst']=data['GoalsAgainst'].apply(pd.to_numeric)
```

```python
data.info()
```

7. List of top 5 teams in terms of points

```python
data['Points']=data['Points'].sort_values(ascending=False)
```

```python
data[['Team','Points']].head()
```

8. function with name “Goal_diff_count” which returns all the teams with their Goal Differences

```python
def Goal_diff_count():
    data['Goal_diff_count']=data['GoalsFor']-data['GoalsAgainst']
    return data[['Team','Goal_diff_count']].sort_values(by='Goal_diff_count',ascending=False)

Goal_diff_count()
```

 11. a new column with name “Winning Percent” and append it to the data set

```python
data['WinningPercent']=(data['GamesWon'].astype(int)/data['GamesPlayed'].astype(int))*100
```

12. If there are any numerical error, replace it with 0%

```python
data['WinningPercent'].fillna(0,inplace=True)
```

13. top 5 teams which has the highest Winning percentage

```python
data[['Team','WinningPercent']].sort_values(by='WinningPercent',ascending=False).head()
```

14. Group teams based on their “Best position” and print the sum of their points for all positions

```python
data[['BestPosition','Points']].groupby('BestPosition').sum().sort_values(by='Points',ascending=False)
```
